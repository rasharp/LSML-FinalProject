# Text preprocessing
1. Delete URL and e-mails
1. Delete numbers, punctuation and stopwords
1. Lower case

```python
import re
import tqdm

import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
STOPWORDS = set(stopwords.words('english'))

from nltk.tokenize import RegexpTokenizer
```

```python
STOPWORDS.add('reuters')
STOPWORDS.add('said')
```

```python
re_url = re.compile(r"https?:\/\/\S+|\S+\.com|\S+.net|\S+.org|www.\S+")
re_mail = re.compile(r"\S+@\S+")

tokenizer = RegexpTokenizer("[A-Za-z]+")
```

```python
def prepare_text(text):
  text = re_mail.sub(' ', re_url.sub(' ', text))
  tokens = tokenizer.tokenize(text.lower())
  return " ".join([w for w in tokens if w not in STOPWORDS])
```

```python
corpus = []
for text in tqdm.tqdm_notebook(df_corpus['TEXT'].dropna()):
  corpus.append(prepare_text(text))
```

# Fit the model

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
tf_vectorizer = CountVectorizer(max_df=0.95, min_df=2,
                                max_features=10000,
                                stop_words='english')
```

```python
data = tf_vectorizer.fit_transform(corpus)
```

```python
lda = LatentDirichletAllocation(n_components=10, max_iter=5,
                                learning_method='online',
                                learning_offset=50.,
                                random_state=0)
```

```python
lda.fit(data)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_top_words(model, feature_names, n_top_words, title):
    fig, axes = plt.subplots(2, 5, figsize=(30, 15), sharex=True)
    axes = axes.flatten()
    for topic_idx, topic in enumerate(model.components_):
        top_features_ind = topic.argsort()[:-n_top_words - 1:-1]
        top_features = [feature_names[i] for i in top_features_ind]
        weights = topic[top_features_ind]

        ax = axes[topic_idx]
        ax.barh(top_features, weights, height=0.7)
        ax.set_title(f'Topic {topic_idx +1}',
                     fontdict={'fontsize': 30})
        ax.invert_yaxis()
        ax.tick_params(axis='both', which='major', labelsize=20)
        for i in 'top right left'.split():
            ax.spines[i].set_visible(False)
        fig.suptitle(title, fontsize=40)

    plt.subplots_adjust(top=0.90, bottom=0.05, wspace=0.90, hspace=0.3)
    plt.show()
```

```python
feature_names = tf_vectorizer.get_feature_names()
plot_top_words(lda, feature_names, 5, 'Topics in LDA model')
```

```python
def get_text_topics(text, model):
  text = prepare_text(text)
  vtext = tf_vectorizer.transform([text])
  res = model.transform(vtext)

  return res


def plot_text_topics(model, vectorizer, text, n_topics=5):
    text_topics = get_text_topics(text, model).flat
    top_topics = np.argsort(text_topics)[:-n_topics-1:-1]
    
    feature_names = tf_vectorizer.get_feature_names()


    fig, axes = plt.subplots(1, n_topics, figsize=(30, 7), sharex=True)
    axes = axes.flatten()
    components = model.components_[top_topics]
    norm_weigh_sum = model.components_.sum(axis=1).reshape(-1, 1)
    for topic_idx, topic in enumerate(components):
        top_features_ind = topic.argsort()[:-6:-1]
        top_features = [feature_names[i] for i in top_features_ind]
        # percent weights
        weights = topic[top_features_ind] / norm_weigh_sum[topic_idx] * 100

        ax = axes[topic_idx]
        ax.barh(top_features, weights, height=0.7)
        ax.set_title(f'Topic {top_topics[topic_idx] +1} - {text_topics[top_topics[topic_idx]]*100:.2f}%',
                     fontdict={'fontsize': 24})
        ax.invert_yaxis()
        ax.tick_params(axis='both', which='major', labelsize=20)
        # for i in 'top right left'.split():
        #     ax.spines[i].set_visible(False)

    plt.subplots_adjust(top=0.90, bottom=0.05, wspace=0.90, hspace=0.3)
    plt.show()


def json_text_topics(model, vectorizer, text, n_topics=5):
    text_topics = get_text_topics(text, model).flat
    top_topics = np.argsort(text_topics)[:-n_topics-1:-1]
    
    feature_names = tf_vectorizer.get_feature_names()

    components = model.components_[top_topics]
    norm_weigh_sum = model.components_.sum(axis=1).reshape(-1, 1)
    res = {}
    for topic_idx, topic in enumerate(components):
        top_features_ind = topic.argsort()[:-6:-1]
        top_features = [feature_names[i] for i in top_features_ind]
        weights = topic[top_features_ind] / norm_weigh_sum[topic_idx] * 100
        
        res_item = {feature: weight for feature, weight in zip(top_features, weights)}
        res["Topic "+str(topic_idx+1)] = (text_topics[top_topics[topic_idx]]*100, res_item)
    
    return json.dumps(res)
```
